```python
csv_airbnb = "/Users/Pere/Desktop/scraper/data/airbnb_data.csv"
csv_listingcounts = "/Users/Pere/Desktop/scraper/data/host_listings_count.csv"

def getAirbnb(dataFile, hostsDataFile):
    data = csv.DictReader(open(dataFile))
    host_listings_count = csv.DictReader(open(hostsDataFile))
    res = {}
    count1 = 0
    for row in data:
        res[row["room_id"]] = {
            "host_id":row["host_id"],
            "room_type":row["room_type"],
            "rating":row["overall_satisfaction"],
            "accommodates":row["accommodates"],
            "bedrooms":row["bedrooms"],
            "bathrooms":row["bathrooms"],
            "price":row["price"],
            "minstay":row["minstay"],
            "property_type":row["property_type"],
            "host_is_superhost":row["host_is_superhost"],
            "amenities":row["amenities"],
            "cancelation_policy":row["cancelation_policy"],
            "latitude":row["latitude"],
            "longitude":row["longitude"],
            "host_listings_count":0
        }
    
    # fill host listings count
    for i in host_listings_count:
        for room_id in res.keys():
            if i["host_id"] == res[room_id]["host_id"]:
                res[room_id]["host_listings_count"] = i["count"]
                break
    return res

airbnb = getAirbnb(csv_airbnb, csv_listingcounts)
```

## Neighbourhoods geojson file of Barcelona

This following code is provided with a geojson file containing each Barcelona's neighbourhood polygon (in coordinates)  which is handled in order to asign each listing to its corresponding neighbourhood. 

```python
# NEIGHBOURHOODS GEOJSON FILE HANDLER

from shapely.geometry import shape, Point

f_geojson = "/Users/Pere/Desktop/scraper/data/neighbourhoods.geojson"

# get json file as js
with open(f_geojson,'r') as f:
    js = json.load(f)
```

```python
# this code below fills up the neighourhood dict, in which contains the neighbourhood for each room_id.
# Note that it takes a long time.

neighbourhoods = {}

for row in airbnb_pred.iterrows():
    neighbourhoods[row[0]] = ""
    lat = row[1]['latitude']
    lon = row[1]['longitude']
    p = Point(lon,lat)
    for feature in js['features']:
        polygon = shape(feature['geometry'])
        if polygon.contains(p):
            neighbourhoods[row[0]] = (feature['properties']['neighbourhood'].encode('ascii','ignore'))
            break
```

```python
# Create new neighbourhoods dataframe to be merged with the airbnb general df

neighs = pd.DataFrame(index = neighbourhoods.keys(), columns=["neighbourhood"], data = neighbourhoods.values())
airbnb_vis["neighbourhood"] = neighs.neighbourhood
airbnb_pred["neighbourhood"] = neighs.neighbourhood
```

Which neighbourhood is more rentable? Is there a correlation between neighourhoods touristic points and listings' rentability?

### 2. Availability

```python
# Getting the data 

csv_calendar = "/Users/Pere/Desktop/scraper/data/calendar_insideairbnb.csv"
csv_reviews = "/Users/Pere/Desktop/scraper/data/reviews_insideairbnb.csv"

calendar = pd.read_csv(csv_calendar)
reviews = pd.read_csv(csv_reviews)
```

```python
calendar = calendar[calendar.available=='t'] #We only want true availability
```

```python
cal_listings = np.unique(calendar["listing_id"].values.ravel())
availability = {}

# Gets each month availability of each listing. This takes a long time to execute.
for listing in cal_listings:
    availability[str(listing)] = {x:0 for x in range(0,13)}
    cal = calendar[calendar['listing_id']==listing]
    for date in cal.date:
        d = datetime.strptime(date , '%Y-%m-%d')
        if d.year == 2016:
            availability[str(listing)][d.month] += 1
```

```python
cols = ['January','February','March','April','May','June','July','August',
       'September','October','November','December']

avail = pd.DataFrame.from_dict(availability).transpose()
avail = avail.drop(avail.columns[0],1)
avail.columns = cols

# Merge each dataset with yielded avail dataframe. Now each listing_id will display its month availability

airbnb_vis2 = pd.merge(airbnb_vis,avail,right_index=True, left_index=True)
airbnb_pred2 = pd.merge(airbnb_pred,avail,right_index=True, left_index=True)
```

```python
yDays = 365
#a = np.array(, dtype='float64')
total = [] # sum of all available days
perct = [] # percentage of total sum over days year
for row in airbnb_pred2.iterrows():
    total.append(float(sum(row[1][cols])))
    perct.append(round(float(float(sum(row[1][cols]))/float(yDays))*100,2)) # sum of all month available days / days year
    
total_np = np.asarray(total)
perct_np = np.asarray(perct)

airbnb_pred2['Total_days'] = total_np
airbnb_pred2['Availability'] = perct_np

airbnb_vis2['Total_days'] = total_np
airbnb_vis2['Availability'] = perct_np
```

### 3. Activity and Occupancy rate

```python
# data -> reviews
#reviews
#Counter(reviews.listing_id)

## Add to general df airbnb2 the mean of the number of each listing's month review counts like:
cols = ['Ja_rcounts','Fe_rcounts','Mar_rcounts','Ap_rcounts','May_rcounts','Jun_rcounts','Jul_rcounts','Au_rcounts',
       'Se_rcounts','Oc_rcounts','No_rcounts','De_rcounts']

review_listings = np.unique(reviews["listing_id"].values.ravel())
reviews_count = {}

years = [2009,2010,2011,2012,2013,2014,2015,2016]

# Gets each month review counts for each listing during all listing's history.
for listing in review_listings:
    reviews_count[str(listing)] = {x:0.0 for x in range(1,13)}
    rev = reviews[reviews['listing_id']==listing]
    month_counts = {i:{year:0 for year in years} for i in range(1,13)} # This list helps get the mean of each month rcounts
    for date in rev.date:
        d = datetime.strptime(date , '%Y-%m-%d')
        month_counts[d.month][d.year] = 1
        reviews_count[str(listing)][d.month] += 1
    for month in reviews_count[str(listing)].keys():
        total_month_counts = reviews_count[str(listing)][month]
        total_year_appearances = sum([month_counts[month][year] for year in years])
        if total_year_appearances is not 0:
            reviews_count[str(listing)][month] = round(total_month_counts / total_year_appearances,2)
        else:
            reviews_count[str(listing)][month] = round(0.0,2)
```

```python
# Creating and merging new dataframes

reviews_df = pd.DataFrame.from_dict(reviews_count).transpose()
reviews_df.columns = cols

airbnb_pred3 = pd.merge(airbnb_pred2,reviews_df,right_index=True, left_index=True)
airbnb_vis3 = pd.merge(airbnb_vis2,reviews_df,right_index=True, left_index=True)
```

```python
a = []
for row in airbnb_pred3.iterrows():
    a.append(sum(row[1][cols]))

total_rcounts = np.asarray(a)
airbnb_pred3['Total_rcounts'] = total_rcounts
airbnb_vis3['Total_rcounts'] = total_rcounts
```

```python
# Estimated bookings

review_rate = .5 # According to InsideAirbnb, 50% of review rate is proper

est_bookings = airbnb_pred3.Total_rcounts * review_rate
airbnb_pred3['Est_bookings'] = est_bookings # Estimated bookings per year
airbnb_vis3['Est_bookings'] = est_bookings # Estimated bookings per year
```

```python
airbnb_vis3.head()
```

```python
# Where no public statements were made about average stays, a minimum stay of 3 nights has been used
avg_minstay = 3

aux = airbnb_pred3.minstay > avg_minstay # Excepcional listing cases, where they have longer minstay than average

stat1 = airbnb_pred3[~aux] # Listings with minstay <= 3
stat2 = airbnb_pred3[aux] # Listings with minstay > 3

occupancy1 = stat1.Est_bookings * avg_minstay
occupancy2 = stat2.Est_bookings * stat2.minstay

occ_rate1 = (occupancy1 / stat1.Total_days) *100
occ_rate2 = (occupancy2 / stat2.Total_days) *100

occ_rate = np.concatenate([occ_rate1,occ_rate2])

# Aquells casos on el rate sigui major a 100 es degut a la debil estimacio dels est. bookings i llavors hi haura 
# habitatges on tendran mes dies d'ocupacio dels disponibles. En aquests casos els hi assignam un 100%
occ_rate[occ_rate > 100] = 100. 
```

```python
airbnb_pred3["Occupancy_rate"] = occ_rate
airbnb_vis3["Occupancy_rate"] = occ_rate
```

### 4. Revenue

Before going farther to the Revenue section, some data cleaning regarded to price is needed. See following boxplot and the outliers sitting in the price column in the dataset.

```python
airbnb_vis3.price.plot(kind='box')
```

Such a big number of outliers are just some excepcional listing's prices on which nothing meaningful could be retrieved in terms of pricing predictability. So keeping those rows with such outliers will only help the model decreasing its predictive performance.

```python
# Outliers elimination
# Keep only the ones that are within +3 to -3 standard deviations in the column price

airbnb_vis3 = airbnb_vis3[np.abs(airbnb_vis3.price-airbnb_vis3.price.mean())<=(3*airbnb_vis3.price.std())]
airbnb_pred3 = airbnb_pred3[np.abs(airbnb_pred3.price-airbnb_pred3.price.mean())<=(3*airbnb_pred3.price.std())]
```

### Starting to make predictions

We will use the following lineal regressors coming from the package lineal_models from Scikit-learn:

#### 1. Linear Regression
#### 2. Ridge
#### 3. Lasso Regression
#### 4. ElasticNets
#### 5. Bayesian Ridge

#### Evaluation
To evaluate which model(s) are doing better, we will need some way to score the results. In this example I've chosen median absolute error, mainly because it makes sense at a glance (it easily translates to a dollar amount relative to price) and is less sensitive to outliers than other metrics like mean squared error.

Speaking of which, it is fairly likely that we have some outliers in the data since we haven't done any filtering or clustering for them, so this is a good way to get a quick and dirty measure of performance before we move to fine-tuning (and of course, we could do more with outliers in the data preparation step).


```python
rs = 1
ests = [ linear_model.LinearRegression(), linear_model.Ridge(),
        linear_model.Lasso(), linear_model.ElasticNet(),
        linear_model.BayesianRidge()]
ests_labels = np.array(['Linear', 'Ridge', 'Lasso', 'ElasticNet', 'BayesRidge'])
errvals = np.array([])

X_train, X_test, y_train, y_test = train_test_split(abnb1.drop(['price'],axis=1),
                                                   abnb1.price, test_size=0.2, random_state=20)

for est in ests:
    est.fit(X_train, y_train)
    err = metrics.median_absolute_error(y_test, est.predict(X_test))
    #print "got error %0.2f" % err
    errvals = np.append(errvals, err)

pos = np.arange(errvals.shape[0])
srt = np.argsort(errvals)
plt.figure(figsize=(7,5))
plt.bar(pos, errvals[srt], align='center')
plt.xticks(pos, ests_labels[srt])
pylab.xlabel('Estimator')
pylab.ylabel('Median Absolute Error (eur)')
```

Apparently, the best result out from those linear regressors is the BayesRidge, being able to predict the price with a median squared error of around 16-18€, which is not a bad but improvable result. Improvable mainly because we have not done yet any tuning of the data and also most of these regressors do not work well with a high-dimensional data (around 100 predictors).

Next we'll try an ensemble method to see if we can get better results. The upside is that we will likely get a better score, and the downside is that the ensemble methods have a bewildering variety of hyperparameters that must be tuned, each of which can affect our model and requires some experimentation to get right. The common way to approach this is to use an exhaustive "grid search" which simply tries all the supplied parameter combinations and uses cross-validation folding to find the best one. Scikit-learn provides the very handy GridSearchCV function for this purpose.

The tradeoff of using GridSearchCV is that the exhaustive search and cross-validation can take a lot of CPU and time. This is where we can use Spark to distribute the search to more machines and cores, enabling us to test more combinations faster.

For our first attempt, we'll limit the number of parameters just so we can get the results back quickly and see if we're doing better than any of the individual methods above.

```python
n_est = 300

tuned_parameters = {
    "n_estimators": [ n_est ],
    "max_depth" : [ 4 ],
    "learning_rate": [ 0.01 ],
    "min_samples_split" : [ 1 ],
    "loss" : [ 'ls', 'lad' ]
}

gbr = ensemble.GradientBoostingRegressor()
clf = GridSearchCV(gbr, cv=3, param_grid=tuned_parameters,
        scoring='median_absolute_error')
preds = clf.fit(X_train, y_train)
best = clf.best_estimator_

metrics.median_absolute_error(y_test, best.predict(X_test))
```

The result of this attempt is a median error of ~12.35€, considerably better outcome than the result of the best regressor deployed before (BayesRidge regressor)! 

Let's get a quick measure of how the error is affected by each round of boosting, so we can see if throwing more iterations at the problem is going to help.

```python
# plot error for each round of boosting
test_score = np.zeros(n_est, dtype=np.float64)

train_score = best.train_score_
for i, y_pred in enumerate(best.staged_predict(X_test)):
    test_score[i] = best.loss_(y_test, y_pred)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(np.arange(n_est), train_score, 'darkblue', label='Training Set Error')
plt.plot(np.arange(n_est), test_score, 'red', label='Test Set Error')
plt.legend(loc='upper right')
pylab.xlabel('Boosting Iterations')
pylab.ylabel('Least Absolute Deviation')
```

It looks like the curve is flattening near the right side of the plot (after around 200-250 iterations), but is still benefitting from more iterations, so we could increase the iterations to 500 without much thought.

```python
n_est = 500

tuned_parameters = {
    "n_estimators": [ n_est ],
    "max_depth" : [ 4 ],
    "learning_rate": [ 0.01 ],
    "min_samples_split" : [ 1 ],
    "loss" : [ 'ls', 'lad' ]
}

gbr = ensemble.GradientBoostingRegressor()
clf = GridSearchCV(gbr, cv=3, param_grid=tuned_parameters,
        scoring='median_absolute_error')
preds = clf.fit(X_train, y_train)
best = clf.best_estimator_

metrics.median_absolute_error(y_test, best.predict(X_test))
```

The error has gone down just a little bit (0.4€) by just adding 200 more iterations. Therefore it does not make much sense to increase iterations on the boosting since the outcome will not be much better. Such a reasoning can be obtained from just paying attention on the test and training least absolute error curve.

## Microneighbourhoods

According to the article written by Airbnb data science team, a microneighbourhood within a city is known as those virtual neighbourhoods which boundaries are defined by the relationships around the price of the listings laying in there. Depending on a listing's location and price, a microneighbourhood could be bigger or smaller based on the similiarities on the other listings' prices and locations. When it comes to differentiate unknown regions in a city, these microneighbourhoods shall help to data scientists to find out which zones could have their listings more expensive or cheaper.

The implemented algorithm is basically a decision tree regressor deployed on a three dimensional space: longitude, latitude and price. Depending on the prices around each point the decision tree should be able to differentiate different regions. Therefore the number of microneighbourhoods are obtained from pruning the yielded decision tree. Also the squares defining each microneighbourhood are obtained by drawing out the tree's features (latitude and longitude).

### Data pre-processing trick

This trick will fill up every single coordinate (lon,lat) that does not have any value with a certain smoothy value (price mean or negative value). This procedure is supposed to help the tree to better classify the microneighbourhood boundaries and also clean up the outlier microneighbourhoods (those that are around the boundaries of Barcelona). 

```python
lons = np.arange(min(dat.longitude),max(dat.longitude),0.01)  # Getting each posible longitude value representing Barcelona 
lats = np.arange(min(dat.latitude),max(dat.latitude),0.01)    # Getting each posible latitude value representing Barcelona
```

```python
# This code takes infinite time to execute

def fill_df(df, value):
    """ Fills up df with value (if not already in there) """
    
    lons = np.arange(min(df.longitude),max(df.longitude),0.001)  # Getting each posible longitude value representing Barcelona 
    lats = np.arange(min(df.latitude),max(df.latitude),0.001)    # Getting each posible latitude value representing Barcelona
    new_rows = []
    for lo in np.nditer(lons):
        for la in np.nditer(lats):
            res = df.isin((np.float(lo), np.float(la)))
            if not ((res['longitude']==True) & (res['latitude']==True)).any(): # If lo,la values not in df
                new_obs = {'longitude':np.float(lo), 'latitude':np.float(la), 'price':value}
                new_rows.append(new_obs)
    
    return pd.DataFrame(new_rows)
```

### Build each microneighbourhood by getting each leaf's decision path

Things to know about first:

1. The output will be comprised of as many microneighbourhoods as many leaves there are. So play out with the `mean_samples_leaf` and `max_depth` decision tree regressor parameters in order to change the number of microneighbourhoods yielded.
2. The microneighbourhoods are attempted to be polygons from which could be useful to create a geojson file. 
3. Each one will be built up by getting the decision tree rules needed to reach such a leaf. See getDecisionPaths and getTreeBoxes methods to know about the implementation.

Geojson library has pretty good tools for dealing with GEOJson Files so it will be used to handle section 2 mentioned above.

```python
from shapely.geometry import *
import geojson
```

```python
def RevTraverseTree(tree, node, rules):
    '''
    Traverase an skl decision tree from a node (presumably a leaf node)
    up to the top, building the decision rules. The rules should be
    input as an empty list, which will be modified in place. The result
    is a nested list of tuples: (feature, direction (left=-1), threshold).  
    The "tree" is a nested list of simplified tree attributes:
    [split feature, split threshold, left node, right node]
    '''
    # now find the node as either a left or right child of something
    # first try to find it as a left node
    try:
        prevnode = tree[2].index(node) #tree[2] means child_left
        leftright = -1
    except ValueError:
        # failed, so find it as a right node
        prevnode = tree[3].index(node) #tree[3] means child_right
        leftright = 1
    # now let's get the rule that caused prevnode to -> node
    rules.append((tree[0][prevnode],leftright,tree[1][prevnode]))
    # if we've not yet reached the top, go up the tree one more step
    if prevnode != 0:
        RevTraverseTree(tree, prevnode, rules)

def getDecisionPaths(tree):
    # get the nodes which are leaves
    leaves = tree.tree_.children_left == -1
    leaves = np.arange(0,dt.tree_.node_count)[leaves]
    
    # build a simpler tree as a nested list: [split feature, split threshold, left node, right node]
    thistree = [dt.tree_.feature.tolist()]
    thistree.append(dt.tree_.threshold.tolist())
    thistree.append(dt.tree_.children_left.tolist())
    thistree.append(dt.tree_.children_right.tolist())
    
    # The returning list will contain tuples in which will be represented each leaf node with its decision_path
    # example: print(decision_paths[0]) # (4, 
    #                                   [(1, -1, 41.40018081665039),
    #                                   (0, 1, 2.109476089477539), 
    #                                   (0, -1, 2.2099030017852783)]
    
    decision_paths = []
    for (ind,nod) in enumerate(leaves):
        # get the decision rules in numeric list form
        rules = []
        RevTraverseTree(thistree, nod, rules)
        #print("----> Rules of leaf node", nod, "are:", rules)
        decision_paths.append((nod,rules))
    
    return decision_paths
```

```python
def getTreeBoxes(tree, feature_names):
    
    leaves = tree.tree_.children_left == -1 
    leaves = np.arange(0,dt.tree_.node_count)[leaves]   # Get the leaves nodes
    values = tree.tree_.value                           # Get all nodes' values
    decision_paths = getDecisionPaths(tree)             # Get decision paths of all leaves
    tree_boxes = []                                     # Final boxes (polygons) yielded by decision tree
                                                        # A polygon follows the following structure:
                                                        # poly = Polygon( ((0, 0), (0, 1), (1, 1), (0, 0)) )
    min_lat = min(dat.latitude)
    min_lon = min(dat.longitude)
    max_lat = max(dat.latitude)
    max_lon = max(dat.longitude)
    
    for (ind,i) in enumerate(leaves):
        lat_up, lat_down, lon_left, lon_right = max_lat, min_lat, min_lon, max_lon
        
        path = decision_paths[ind][1]                   # decision path of the i-th leaf
        value = round(float(values[i]),2)               # price predicted in the i-th leaf
        
        for idx in range(len(path)-1, -1, -1):          # going backwards on the list means going from firsts 
                                                        #thresholds to final ones
            if feature_names[path[idx][0]] == 'longitude':
                if path[idx][1] == -1:                  # went left
                    lon_right = path[idx][2]
                else:                                   # went right
                    lon_left = path[idx][2]
            else:
                if path[idx][1] == -1:                  # went left (predicted region downwards)
                    lat_up = path[idx][2]
                else:                                   # went right (upwards)
                    lat_down = path[idx][2]
        
        # Polygon's construction
        b = ((lon_left,lat_up),(lon_left,lat_down),(lon_right,lat_down),(lon_right,lat_up),(lon_left,lat_up))   
        tree_boxes.append([b,value])
    
    return tree_boxes
```

#### Getting all the yielded polygons/microneighbourhoods and converting them into a unique geojson file

```python
def write_json(newfile, fc):
    #fc is a geojson Feature Collection object
    with open(newfile, 'w') as outfile:
        geojson.dump(fc, outfile)
```

```python
boxes = getTreeBoxes(dt, feature_names)

features = []
for b in boxes:
    if b[1] > 10.0: #Eliminam aquells poligons on el valor és menor que un cert valor molt baix (proper a 0)
        poly = geojson.Polygon([b[0]])
        feature = geojson.Feature(geometry=poly, properties={"price":b[1]})
        features.append(feature)
    

fc = geojson.FeatureCollection(features)

newfile = "/Users/Pere/Desktop/scraper/data/trick2_minleaf100_depth20.geojson"

write_json(newfile, fc)
```
